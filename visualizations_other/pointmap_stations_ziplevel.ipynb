```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import plotly.express as px
import seaborn as sns
import geopandas as gpd

plt.style.use('ggplot')
pd.set_option('display.max_columns', 30)
```

```python
# import trip data
df = pd.read_csv('data/citibike_ZIP.csv')
df.shape
```

```python
# convert to geo dataframe
start_gpd = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(
    df['start_lon'], df['start_lat']), crs = 'epsg:4326')

# update projection 
start_gpd.to_crs(epsg=2263, inplace = True)
```

```python
# import zipcode shapefile
zip_shp = gpd.read_file('nyc_zip/ZIP_CODE_040114.shp')[['ZIPCODE', 'geometry']]
```

```python
start_count = df.groupby(['start_id'], as_index=False)['bikeid'].count()
start_count.columns = ['start_id','start_count']
start_count.head()
```

```python
# drop duplicates for plotting
gpd_sub = start_gpd.drop_duplicates(['start_id'])
gpd_sub.shape
```

```python
gpd_sub = gpd_sub.merge(start_count,
                       on = 'start_id',
                       how = 'left')
```

```python
man_zip = zip_shp.loc[zip_shp['ZIPCODE'] == '10003']
bk_zip = zip_shp.loc[zip_shp['ZIPCODE'] == '11201']

#man_gpd = gpd_sub.loc[gpd_sub['ZIPCODE'] == 10003]
#bk_gpd = gpd_sub.loc[gpd_sub['ZIPCODE'] == 11201]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, 
                        figsize=(20, 10),
                        sharex=False,
                        sharey=False,
                        constrained_layout=True)

fig.suptitle('Trips Originating in the East Village and DUMBO', fontsize=20)

zip_shp.plot(ax=ax1, color = 'lightgrey', edgecolor = 'silver', linewidth = 2)
man_zip.plot(ax=ax1, color='dodgerblue', edgecolor = 'black', linewidth = 2)
gpd_sub.plot(ax=ax1, markersize = gpd_sub['start_count']/5, color = 'darkslategray')

minx, miny, maxx, maxy = man_zip.total_bounds
ax1.set_xlim(minx - 750, maxx + 750)
ax1.set_ylim(miny - 100, maxy + 100)

ax1.set_title('10003', pad = 15)
ax1.set_facecolor("whitesmoke")
ax1.grid(False)
ax1.set_xticks([])
ax1.set_yticks([])

zip_shp.plot(ax=ax2, color = 'lightgrey', edgecolor = 'silver', linewidth = 2)
bk_zip.plot(ax=ax2, color='orchid', edgecolor = 'black', linewidth = 2)
gpd_sub.plot(ax=ax2, markersize = gpd_sub['start_count']/8, color = 'darkslategray')

minx, miny, maxx, maxy = bk_zip.total_bounds
ax2.set_xlim(minx - 100, maxx + 100)
ax2.set_ylim(miny - 750, maxy + 750)

ax2.set_title('11201', pad = 15)
ax2.set_facecolor("whitesmoke")
ax2.grid(False)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.legend(['Monthly Trips'],loc='upper right')
```
